matplotlib
==========

We will uise matplotlib to use graphs. It has we organized documentation [here](http://matplotlib.org/). 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

plot
----

```python

x = np.arange(0, 2* np.pi, 0.1)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine')
plt.legend(['sin','cos'])
plt.show()
```

subplot
--------

```python
x = np.arange(0, 2* np.pi, 0.1)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.title('Sine and Cosine')
plt.ylabel('y')
plt.subplot(2,1,2)
plt.plot(x, np.cos(x),'|')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
